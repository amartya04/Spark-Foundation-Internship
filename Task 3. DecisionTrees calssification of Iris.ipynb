```python
# Importing libraries in Python
import sklearn.datasets as datasets
import pandas as pd


# Loading the iris dataset
iris=datasets.load_iris()
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import tree

# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

### Now let us define the Decision Tree Algorithm

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)

print('Decision Tree Classifer Created')
```

```python
feature_n=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
class_n=['setosa','versicolor','virginica']
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)
tree.plot_tree(dtree,feature_names=feature_n,class_names=class_n,filled='True',rounded='True')
```

```python
# Import necessary libraries for graph viz
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
